```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
  return 3*x**2 - 4*x + 5
```

```python
f(3)
```

```python
xs = np.arange(-5,5,0.25)
ys = f(xs)
plt.plot(xs,ys)
```

```python
def derivative(x,xh,h):
  return (xh - x)/h
```

```python
h= 0.0000001
x = 3
xh = f(x+h)
print(xh)
print(derivative(f(x),xh,h))
```

```python
slope = derivative(xs,ys,h)
```

```python
fig = plt.figure()
# plotting
plt.plot(xs, slope)
plt.show()
```

```python
class Value:
  def __init__(self, data , _children=(), _op="" ):
    self.data = data
    self._prev = set(_children)
    self._op = _op

  def __repr__(self) -> str:
    return f"Value(data={self.data})"

  def __add__(self,other):
    output = Value(self.data + other.data, (self,other), "+")
    return output

  def __mul__(self,other):
    output = Value(self.data * other.data, (self,other), "*")
    return output

a = Value(2)
b = Value(-3)
c = Value(10)
d = a * b + c
d._prev
d._op
```
